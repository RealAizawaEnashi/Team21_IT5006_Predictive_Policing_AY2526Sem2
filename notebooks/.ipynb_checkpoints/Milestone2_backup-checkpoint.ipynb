```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
```

```python
BASE = "https://data.cityofchicago.org/resource/ijzp-q8t2.json"

COLUMNS = [
    "id",
    "date",
    "primary_type",
    "arrest",
    "domestic",
    "district"
]

SELECT_COLS = ",".join(COLUMNS)

START_DATE = "2015-01-01T00:00:00"
END_DATE   = "2025-12-31T23:59:59"
```

```python
def fetch_chicago_crime(base_url, select_cols, start_date, end_date, batch_size=50000, max_rows=None):
    all_rows = []
    offset = 0

    while True:
        params = {
            "$select": select_cols,
            "$where": f"date >= '{start_date}' AND date <= '{end_date}'",
            "$limit": batch_size,
            "$offset": offset,
            "$order": "date ASC"
        }

        r = requests.get(base_url, params=params, timeout=60)
        r.raise_for_status()
        batch = r.json()

        if not batch:
            print(f"Finished. Total rows fetched: {offset}")
            break

        all_rows.extend(batch)
        offset += len(batch)

        print(f"Fetched {offset} rows...")

        if max_rows is not None and offset >= max_rows:
            print(f"Stopped early at {offset} rows (max_rows reached).")
            break

    df = pd.DataFrame(all_rows)
    return df
```

```python
df_full = fetch_chicago_crime(
    base_url=BASE,
    select_cols=SELECT_COLS,
    start_date=START_DATE,
    end_date=END_DATE,
    batch_size=50000,
    max_rows=None
)
```

```python
# 数据清洗

df = df_full.copy()

# 时间字段
df["date"] = pd.to_datetime(df["date"], errors="coerce")

# district 转数值
df["district"] = pd.to_numeric(df["district"], errors="coerce")

# 布尔字段转 0/1
df["arrest"] = df["arrest"].astype(str).str.lower().map({"true": 1, "false": 0})
df["domestic"] = df["domestic"].astype(str).str.lower().map({"true": 1, "false": 0})

# 文本字段统一格式
df["primary_type"] = df["primary_type"].astype(str).str.strip().str.upper()

# 去重
df = df.drop_duplicates(subset=["id"])

# 去掉关键缺失
df = df.dropna(subset=["id", "date", "district", "primary_type"])

print(df.shape)
df.head()
```

```python
df.info()
df[["id", "date", "district", "primary_type", "arrest", "domestic"]].head()
```

```python
# 做时间特征
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day_of_week"] = df["date"].dt.dayofweek
df["hour"] = df["date"].dt.hour
df["is_weekend"] = (df["day_of_week"] >= 5).astype(int)

# 每条记录所属周的起始时间
df["week_start"] = df["date"].dt.to_period("W").apply(lambda r: r.start_time)
```

```python
df[["date", "week_start", "year", "month", "day_of_week", "hour", "is_weekend"]].head()
```

```python
# 按 district + week_start 聚合
weekly = (
    df.groupby(["district", "week_start"], as_index=False)
      .agg(
          crime_count=("id", "count"),
          arrest_rate=("arrest", "mean"),
          domestic_rate=("domestic", "mean"),
          theft_count=("primary_type", lambda x: (x == "THEFT").sum()),
          battery_count=("primary_type", lambda x: (x == "BATTERY").sum()),
          burglary_count=("primary_type", lambda x: (x == "BURGLARY").sum()),
          robbery_count=("primary_type", lambda x: (x == "ROBBERY").sum()),
          assault_count=("primary_type", lambda x: (x == "ASSAULT").sum()),
          weekend_ratio=("is_weekend", "mean"),
          night_ratio=("hour", lambda x: ((x >= 20) | (x <= 5)).mean())
      )
)

print(weekly.shape)
weekly.head()
```

```python
#定义hotspot标签
weekly["weekly_rank_pct"] = (
    weekly.groupby("week_start")["crime_count"]
          .rank(method="min", pct=True, ascending=False)
)

weekly["hotspot"] = (weekly["weekly_rank_pct"] <= 0.20).astype(int)
```

```python
weekly[["district", "week_start", "crime_count", "weekly_rank_pct", "hotspot"]].head(20)
```

```python
# 把目标变成“下一周是否 hotspot”
weekly = weekly.sort_values(["district", "week_start"]).reset_index(drop=True)

weekly["target_next_week_hotspot"] = (
    weekly.groupby("district")["hotspot"].shift(-1)
)

weekly = weekly.dropna(subset=["target_next_week_hotspot"])
weekly["target_next_week_hotspot"] = weekly["target_next_week_hotspot"].astype(int)

weekly[["district", "week_start", "hotspot", "target_next_week_hotspot"]].head(20)
```

```python
# lag 特征
weekly = weekly.sort_values(["district", "week_start"]).reset_index(drop=True)

group = weekly.groupby("district")

weekly["lag1_crime_count"] = group["crime_count"].shift(1)
weekly["lag2_crime_count"] = group["crime_count"].shift(2)
weekly["lag3_crime_count"] = group["crime_count"].shift(3)

weekly["lag1_arrest_rate"] = group["arrest_rate"].shift(1)
weekly["lag1_domestic_rate"] = group["domestic_rate"].shift(1)

weekly["rolling4_crime_mean"] = (
    group["crime_count"].shift(1).rolling(4).mean().reset_index(level=0, drop=True)
)

weekly["rolling4_crime_sum"] = (
    group["crime_count"].shift(1).rolling(4).sum().reset_index(level=0, drop=True)
)

for col in ["theft_count", "battery_count", "burglary_count", "robbery_count", "assault_count"]:
    weekly[f"lag1_{col}"] = group[col].shift(1)
```

```python
# === 新增变化率特征：加在 lag / rolling 特征之后 ===

weekly["crime_count_change_1"] = weekly["crime_count"] - weekly["lag1_crime_count"]

weekly["crime_count_growth_1"] = (
    (weekly["crime_count"] - weekly["lag1_crime_count"]) /
    (weekly["lag1_crime_count"] + 1e-6)
)

weekly["crime_count_vs_rolling4"] = (
    weekly["crime_count"] / (weekly["rolling4_crime_mean"] + 1e-6)
)

# 如果你已经有这些 lag 列，也建议一起加
if "lag1_battery_count" in weekly.columns:
    weekly["battery_change_1"] = weekly["battery_count"] - weekly["lag1_battery_count"]

if "lag1_assault_count" in weekly.columns:
    weekly["assault_change_1"] = weekly["assault_count"] - weekly["lag1_assault_count"]

if "lag1_theft_count" in weekly.columns:
    weekly["theft_change_1"] = weekly["theft_count"] - weekly["lag1_theft_count"]
```

```python
# 开始数据建模
# 第一步：先确认标签分布
print("train label ratio:")
print(train_df["target_next_week_hotspot"].value_counts())
print(train_df["target_next_week_hotspot"].value_counts(normalize=True))

print("\nvalid label ratio:")
print(valid_df["target_next_week_hotspot"].value_counts())
print(valid_df["target_next_week_hotspot"].value_counts(normalize=True))

print("\ntest label ratio:")
print(test_df["target_next_week_hotspot"].value_counts())
print(test_df["target_next_week_hotspot"].value_counts(normalize=True))
```

```python
# 定义特征列
feature_cols = [
    "crime_count",
    "arrest_rate",
    "domestic_rate",
    "theft_count",
    "battery_count",
    "burglary_count",
    "robbery_count",
    "assault_count",
    "weekend_ratio",
    "night_ratio",
    "lag1_crime_count",
    "lag2_crime_count",
    "lag3_crime_count",
    "lag1_arrest_rate",
    "lag1_domestic_rate",
    "rolling4_crime_mean",
    "rolling4_crime_sum",
    "lag1_theft_count",
    "lag1_battery_count",
    "lag1_burglary_count",
    "lag1_robbery_count",
    "lag1_assault_count",
    "crime_count_change_1",
    "crime_count_growth_1",
    "crime_count_vs_rolling4",
    "battery_change_1",
    "assault_change_1",
    "theft_change_1"
]
```

```python
# 切 X 和 y
X_train = train_df[feature_cols].copy()
y_train = train_df["target_next_week_hotspot"].copy()

X_valid = valid_df[feature_cols].copy()
y_valid = valid_df["target_next_week_hotspot"].copy()

X_test = test_df[feature_cols].copy()
y_test = test_df["target_next_week_hotspot"].copy()
```

```python
#检查
print(X_train.shape, y_train.shape)
print(X_valid.shape, y_valid.shape)
print(X_test.shape, y_test.shape)
```

```python
# 先跑第一个 baseline：Logistic Regression
log_model = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42))
])

log_model.fit(X_train, y_train)

valid_pred_log = log_model.predict(X_valid)
valid_prob_log = log_model.predict_proba(X_valid)[:, 1]

print("=== Logistic Regression (Validation) ===")
print(classification_report(y_valid, valid_pred_log))
print("ROC-AUC:", roc_auc_score(y_valid, valid_prob_log))
print("Confusion Matrix:")
print(confusion_matrix(y_valid, valid_pred_log))
```

```python
# 再跑第二个模型：Random Forest
from sklearn.ensemble import RandomForestClassifier

rf_model = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("clf", RandomForestClassifier(
        n_estimators=300,
        max_depth=10,
        min_samples_split=10,
        min_samples_leaf=5,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    ))
])

rf_model.fit(X_train, y_train)

valid_pred_rf = rf_model.predict(X_valid)
valid_prob_rf = rf_model.predict_proba(X_valid)[:, 1]

print("=== Random Forest (Validation) ===")
print(classification_report(y_valid, valid_pred_rf))
print("ROC-AUC:", roc_auc_score(y_valid, valid_prob_rf))
print("Confusion Matrix:")
print(confusion_matrix(y_valid, valid_pred_rf))
```

```python
from sklearn.metrics import f1_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

param_grid = [
    {"n_estimators": 100, "learning_rate": 0.03, "max_depth": 2},
    {"n_estimators": 200, "learning_rate": 0.03, "max_depth": 2},
    {"n_estimators": 300, "learning_rate": 0.03, "max_depth": 2},
    {"n_estimators": 200, "learning_rate": 0.05, "max_depth": 3},
    {"n_estimators": 300, "learning_rate": 0.05, "max_depth": 3},
    {"n_estimators": 400, "learning_rate": 0.03, "max_depth": 3},
]

best_score = -1
best_cfg = None
best_model = None

for cfg in param_grid:
    gb_model = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("clf", GradientBoostingClassifier(
            n_estimators=cfg["n_estimators"],
            learning_rate=cfg["learning_rate"],
            max_depth=cfg["max_depth"],
            random_state=42
        ))
    ])
    gb_model.fit(X_train, y_train)
    prob = gb_model.predict_proba(X_valid)[:, 1]
    
    # 顺便一起调阈值
    best_local_f1 = -1
    best_local_t = 0.5
    
    for t in np.arange(0.1, 0.91, 0.02):
        pred = (prob >= t).astype(int)
        f1 = f1_score(y_valid, pred)
        if f1 > best_local_f1:
            best_local_f1 = f1
            best_local_t = t
    
    if best_local_f1 > best_score:
        best_score = best_local_f1
        best_cfg = {**cfg, "threshold": best_local_t}
        best_model = gb_model

valid_prob_gb = best_model.predict_proba(X_valid)[:, 1]
valid_pred_gb = (valid_prob_gb >= best_cfg["threshold"]).astype(int)

print(best_cfg, best_score)
```

```python
# 把三个模型的结果整理成表
from sklearn.metrics import precision_score, recall_score, f1_score

results = []

def collect_metrics(name, y_true, y_pred, y_prob):
    results.append({
        "Model": name,
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1": f1_score(y_true, y_pred),
        "ROC_AUC": roc_auc_score(y_true, y_prob)
    })

collect_metrics("Logistic Regression", y_valid, valid_pred_log, valid_prob_log)
collect_metrics("Random Forest", y_valid, valid_pred_rf, valid_prob_rf)
collect_metrics("Gradient Boosting", y_valid, valid_pred_gb, valid_prob_gb)

results_df = pd.DataFrame(results).sort_values("F1", ascending=False)
results_df
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, f1_score, precision_score, recall_score

# 直接用调参阶段保存下来的最优模型和最优阈值
final_model = best_model
final_threshold = float(best_cfg["threshold"])

test_eval = test_df.copy()
test_eval["pred_prob"] = final_model.predict_proba(X_test)[:, 1]

def mark_top20_as_hotspot(group):
    n = len(group)
    k = max(1, int(np.ceil(n * 0.20)))   # 每周前20%
    top_idx = group["pred_prob"].nlargest(k).index
    group["pred"] = 0
    group.loc[top_idx, "pred"] = 1
    return group

test_eval = (
    test_eval.groupby("week_start", group_keys=False)
    .apply(mark_top20_as_hotspot)
)
test_pred = test_eval["pred"].values
test_prob = test_eval["pred_prob"].values

print("=== Tuned Gradient Boosting (Test) ===")
print("Best config:", best_cfg)
print(classification_report(y_test, test_pred))
print("F1:", f1_score(y_test, test_pred))
print("Precision:", precision_score(y_test, test_pred))
print("Recall:", recall_score(y_test, test_pred))
print("ROC-AUC:", roc_auc_score(y_test, test_prob))
print("Confusion Matrix:")
print(confusion_matrix(y_test, test_pred))
```

```python
# 做一个简单 spatial evaluation
# 先做一列预测：
test_eval["pred_hotspot"] = test_eval["pred"]
test_eval["true_hotspot"] = test_eval["target_next_week_hotspot"]
```

```python
# 然后按周算命中率：
def weekly_hit_rate(group):
    pred_set = set(group.loc[group["pred_hotspot"] == 1, "district"])
    true_set = set(group.loc[group["true_hotspot"] == 1, "district"])
    
    if len(true_set) == 0:
        return np.nan
    
    return len(pred_set & true_set) / len(true_set)

weekly_spatial_hit = (
    test_eval.groupby("week_start")
    .apply(weekly_hit_rate)
    .reset_index(name="spatial_hit_rate")
)

weekly_spatial_hit.head()
print("Average spatial hit rate:", weekly_spatial_hit["spatial_hit_rate"].mean())
```

```python
# Cross-validation result
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
import numpy as np

def evaluate_with_threshold(y_true, prob, threshold=0.46):
    pred = (prob >= threshold).astype(int)
    return {
        "Precision": precision_score(y_true, pred, zero_division=0),
        "Recall": recall_score(y_true, pred, zero_division=0),
        "F1": f1_score(y_true, pred, zero_division=0),
        "ROC_AUC": roc_auc_score(y_true, prob)
    }
```

```python
# 定义 rolling CV folds
cv_df = weekly.copy()
cv_df["year"] = cv_df["week_start"].dt.year

cv_folds = [
    {"train_years": list(range(2015, 2020)), "valid_year": 2020},
    {"train_years": list(range(2015, 2021)), "valid_year": 2021},
    {"train_years": list(range(2015, 2022)), "valid_year": 2022},
    {"train_years": list(range(2015, 2023)), "valid_year": 2023},
]
```

```python
# 跑每一折
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

cv_results = []

for i, fold in enumerate(cv_folds, start=1):
    train_part = cv_df[cv_df["year"].isin(fold["train_years"])].copy()
    valid_part = cv_df[cv_df["year"] == fold["valid_year"]].copy()

    X_tr = train_part[feature_cols]
    y_tr = train_part["target_next_week_hotspot"]

    X_va = valid_part[feature_cols]
    y_va = valid_part["target_next_week_hotspot"]

    model = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("clf", GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.03,
            max_depth=2,
            random_state=42
        ))
    ])

    model.fit(X_tr, y_tr)

    prob_va = model.predict_proba(X_va)[:, 1]

    # 先用你现在的最终阈值；如果你后面改成 top20% 规则，这里我再带你改
    metrics = evaluate_with_threshold(y_va, prob_va, threshold=0.46)

    cv_results.append({
        "Fold": i,
        "Train_Years": f"{min(fold['train_years'])}-{max(fold['train_years'])}",
        "Valid_Year": fold["valid_year"],
        **metrics
    })

cv_results_df = pd.DataFrame(cv_results)
cv_results_df
```

```python
# mean ± std 汇总表
cv_summary = pd.DataFrame({
    "Metric": ["Precision", "Recall", "F1", "ROC_AUC"],
    "Mean": [
        cv_results_df["Precision"].mean(),
        cv_results_df["Recall"].mean(),
        cv_results_df["F1"].mean(),
        cv_results_df["ROC_AUC"].mean(),
    ],
    "Std": [
        cv_results_df["Precision"].std(),
        cv_results_df["Recall"].std(),
        cv_results_df["F1"].std(),
        cv_results_df["ROC_AUC"].std(),
    ]
})

cv_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
plt.plot(cv_results_df["Valid_Year"], cv_results_df["Precision"], marker="o", label="Precision")
plt.plot(cv_results_df["Valid_Year"], cv_results_df["Recall"], marker="o", label="Recall")
plt.plot(cv_results_df["Valid_Year"], cv_results_df["F1"], marker="o", label="F1")
plt.plot(cv_results_df["Valid_Year"], cv_results_df["ROC_AUC"], marker="o", label="ROC-AUC")

plt.title("Rolling Cross-Validation Performance by Validation Year")
plt.xlabel("Validation Year")
plt.ylabel("Score")
plt.xticks(cv_results_df["Valid_Year"])
plt.ylim(0, 1.05)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
